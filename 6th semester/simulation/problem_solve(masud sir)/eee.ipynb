```python
import numpy as np
import matplotlib.pyplot as plt


central_mean = 100

means = [25, 50, 75, 100, 150, 200, 300]

num_samples = 10000 
plt.figure(figsize=(12, 7))


for m in means:
   
    samples = np.random.exponential(scale=m, size=num_samples)
    
    
    plt.hist(samples, bins=100, density=True, alpha=0.4, label=f"Mean={m} days")


plt.title("Exponential Distributions: 3 Means Before & After Mean=100", fontsize=14)
plt.xlabel("Days until next wave", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.xlim(0, 600)  
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()
```
