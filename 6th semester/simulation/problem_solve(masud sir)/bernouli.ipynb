```python
import numpy as np
import matplotlib.pyplot as plt


means = [25, 50, 75, 100, 150, 200, 300]


probabilities = []

for mean in means:
    lam = 1 / mean          
    x = mean                
    pdf = lam * np.exp(-lam * x)
    probabilities.append(pdf)


plt.figure()
plt.plot(means, probabilities, marker='o')
plt.xlabel("Mean Value")
plt.ylabel("Probability Density at x = mean")
plt.title("Exponential PDF at x = Mean")
plt.grid(True)
plt.show()
```
